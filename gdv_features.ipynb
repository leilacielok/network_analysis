```python
G_und = nx.read_edgelist("G_und.edgelist", nodetype=int)
nodes_ml = pd.read_csv("nodes_ml.csv").iloc[:,0].tolist()

print(G_und.number_of_nodes(), G_und.number_of_edges())
```

```python
# GDV
import orbit_count

node_list = list(G_und.nodes())

gdv = orbit_count.node_orbit_counts(
    G_und,
    graphlet_size=4,
    node_list=node_list
)
```

```python
gdv_dict = {node: gdv[i] for i, node in enumerate(node_list)}

X_gdv = [gdv_dict[n] for n in nodes_ml]

gdv_df = pd.DataFrame(X_gdv, index=nodes_ml)
```

```python
gdv_df.to_csv("gdv_features.csv")
```
